```python
import pandas as pd
```

```python
df = pd.read_csv('한국_기업문화_HR_데이터셋_샘플.csv', encoding='utf-8')
```

```python
df.head()
```

```python
missing_ratio = df.isnull().sum()
missing_ratio = missing_ratio[missing_ratio > 1].round(1)
missing_ratio
```

```python
korea_job_df1 = df[['Age', '이직여부', '집까지거리', '근무환경만족도', '업무만족도', '이전회사경험수', '대인관계만족도']]
```

```python
tmp = []
for each in korea_job_df1['이직여부']:
    tmp.append(0 if each == 'No' else 1)

korea_job_df1['이직여부'] = tmp
```

```python
korea_job_df1
```

```python
tmp = []
for each in korea_job_df1['집까지거리']:
    tmp.append('danger' if each >= 15 else 'okay')

korea_job_df1['집까지거리'] = tmp

korea_job_df1
```

```python
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['근무환경만족도']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df1['근무환경만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df1['근무환경만족도'] = tmp

korea_job_df1
```

```python
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['업무만족도']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df1['업무만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df1['업무만족도'] = tmp

korea_job_df1
```

```python
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['이전회사경험수']
    count += 1
print(a/count)
```

```python
korea_job_df2 = korea_job_df1[['Age', '이직여부', '집까지거리', '근무환경만족도', '업무만족도', '대인관계만족도']]
```

```python
korea_job_df2
```

```python
a=0
count=0
for index, row in korea_job_df2.iterrows():
  if row['이직여부'] == 1:
    a += row['대인관계만족도']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df2['대인관계만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df2['대인관계만족도'] = tmp

korea_job_df2
```

```python
a=0
count=0
for index, row in korea_job_df2.iterrows():
  if row['이직여부'] == 1:
    a += row['Age']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df2['Age']:
    tmp.append('danger' if each <= 35 else 'okay')

korea_job_df2['Age'] = tmp

korea_job_df2
```
