```python
import pandas as pd
```

```python
#데이터 로드
df = pd.read_csv('한국_기업문화_HR_데이터셋_샘플.csv', encoding='utf-8')
```

```python
df.head()
```

```python
#결측치 비율 확인, 결측치 없으므로 결측치 처리 건너뜀
missing_ratio = df.isnull().sum()
missing_ratio = missing_ratio[missing_ratio > 1].round(1)
missing_ratio
```

```python
#중요 컬럼들만 추리기
korea_job_df1 = df[['Age', '이직여부', '집까지거리', '근무환경만족도', '업무만족도', '이전회사경험수', '대인관계만족도']]
```

```python
#이직여부 0,1 데이터로 변환
tmp = []
for each in korea_job_df1['이직여부']:
    tmp.append(0 if each == 'No' else 1)

korea_job_df1['이직여부'] = tmp
```

```python
korea_job_df1
```

```python
#이직한 직원들의 집까지거리 평균
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['집까지거리']
    count += 1
print(a/count)
```

```python
#집까지 거리에 따라 이직위험 여부로 변환
tmp = []
for each in korea_job_df1['집까지거리']:
    tmp.append('danger' if each >= 10 else 'okay')

korea_job_df1['집까지거리'] = tmp

korea_job_df1
```

```python
#이직한 직원들의 근무환경만족도 평균
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['근무환경만족도']
    count += 1
print(a/count)
```

```python
#근무환경만족도에 따라 이직위험 여부로 변환
tmp = []
for each in korea_job_df1['근무환경만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df1['근무환경만족도'] = tmp

korea_job_df1
```

```python
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['업무만족도']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df1['업무만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df1['업무만족도'] = tmp

korea_job_df1
```

```python
#이직한 직원들의 이전회사경험수 평균
a=0
count=0
for index, row in korea_job_df1.iterrows():
  if row['이직여부'] == 1:
    a += row['이전회사경험수']
    count += 1
print(a/count)
```

```python
#큰 상관 없는걸로 보아 컬럼 제거
korea_job_df2 = korea_job_df1[['Age', '이직여부', '집까지거리', '근무환경만족도', '업무만족도', '대인관계만족도']]
```

```python
korea_job_df2
```

```python
a=0
count=0
for index, row in korea_job_df2.iterrows():
  if row['이직여부'] == 1:
    a += row['대인관계만족도']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df2['대인관계만족도']:
    tmp.append('danger' if each <= 2 else 'okay')

korea_job_df2['대인관계만족도'] = tmp

korea_job_df2
```

```python
a=0
count=0
for index, row in korea_job_df2.iterrows():
  if row['이직여부'] == 1:
    a += row['Age']
    count += 1
print(a/count)
```

```python
tmp = []
for each in korea_job_df2['Age']:
    tmp.append('danger' if each <= 35 else 'okay')

korea_job_df2['Age'] = tmp

korea_job_df2
```

```python
korea_job_df2['score']=0
korea_job_df2['prediction']=None

korea_job_df_score = korea_job_df2[['Age', '이직여부', '집까지거리', '근무환경만족도', '업무만족도', '대인관계만족도', 'score', 'prediction']]
korea_job_df_score
```

```python
#의도대로 코딩되지 않음 시간 부족

for each, row in korea_job_df_score.iterrows():
  count=0
  if row['Age'] == 'danger':
    count += 1
  if row['집까지거리'] == 'danger':
    count += 1
  if row['근무환경만족도'] == 'danger':
    count += 1
  if row['업무만족도'] == 'danger':
    count += 1
  if row['대인관계만족도'] == 'danger':
    count += 1
  korea_job_df_score['score'] = count

korea_job_df_score
```
